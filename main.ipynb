## Math Function Graphics

```python
# Imports
import matplotlib.pyplot as plt
from numpy import arange
```

```python
def plot_creator(function,x_range):
    fig, ax = plt.subplots()
    ax.axis("off")
    y_axis = [function(val)*500 for val in x]
    y3 = [0 for _ in x]
    y4 = [i for i in range(int(min(y_axis))-100,int(max(y_axis))+100)]
    zero_x = [0 for _ in range(int(min(y_axis))-100,int(max(y_axis)+100))]
    plt.plot(x,y3,color="black")
    plt.plot(zero_x,y4,color="black")
    
    
    plt.plot(x_range,y_axis)
    return plt
```

```python
# y = kx
x = arange(-1000,1000,1)

def linear(k=1):
    def func_body(x):
        return k*x
    return func_body

lin_plt = plot_creator(linear(k=1),x_range=x)
lin_plt.show()
```

```python
# y = kx^2+b
x = arange(-10,10,1)

def quadratic(k=1,b=0):
    def func_body(x):
        return k*(x**2)+b
    return func_body

quad_plot = plot_creator(quadratic(k=10,b=-100),x_range=x)
quad_plot.show()
```

```python
# y = sin(x)
from math import pi,sin
x = arange(-360,360,1)

def sinusoid(x):
    return sin(x*pi/180)


sin_plot = plot_creator(sinusoid,x_range=x)
sin_plot.show()
```

```python
# y = ln(x)
from math import log
x = arange(1,1000,1)

def logarithmic(x):
    return log(x)

log_plot = plot_creator(logarithmic,x_range=x)
log_plot.show()
```
